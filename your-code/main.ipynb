```python
# Method 1:Using subplots and plotting individually
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))

# Plot (x, y)
axes[0].plot(x, y, color='blue', linewidth=2, linestyle='--')
axes[0].set_title('(x, y)')

# Plot (x, z)
axes[1].plot(x, z, color='red', linewidth=2, linestyle='-')
axes[1].set_title('(x, z)')

plt.show()
```

#### Use plt.subplots(nrows=1, ncols=2) to create the plot.

```python
# Method 2:  Using subplots and plotting together
fig, ax = plt.subplots(figsize=(10, 4))

# Plot (x, y)
ax.plot(x, y, label='y = x * 2', color='blue', linewidth=2, linestyle='--')

# Plot (x, z)
ax.plot(x, z, label='z = x^2', color='red', linewidth=2, linestyle='-')

ax.set_title('(x, y) and (x, z)')
ax.legend()

plt.show()
```

#### Plot both $y=x^2$ and $y=e^x$ in the same plot using normal and logarithmic scale.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
# your code here

# Data
x = np.linspace(0, 5, 100)
y1 = x ** 2
y2 = np.exp(x)


# Create the plot
plt.figure(figsize=(8, 6))

# Plot y=x^2 in normal scale
plt.plot(x, y1, label=r'$y = x^2$', color='blue')

# Plot y=e^x in normal scale
plt.plot(x, y2, label=r'$y = e^x$', color='green')

# Set labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of $y=x^2$ and $y=e^x$')

# Add legend
plt.legend()

# Show the plot in normal scale
plt.show()

# Create the plot with logarithmic y-scale
plt.figure(figsize=(8, 6))

# Plot y=x^2 in logarithmic scale
plt.plot(x, y1, label=r'$y = x^2$', color='blue')

# Plot y=e^x in logarithmic scale
plt.plot(x, y2, label=r'$y = e^x$', color='green')

# Set labels and title
plt.xlabel('x')
plt.ylabel('y (logarithmic scale)')
plt.title('Plot of $y=x^2$ and $y=e^x$ (logarithmic scale)')

# Set y-axis to logarithmic scale
plt.yscale('log')

# Add legend
plt.legend()

# Show the plot in logarithmic scale
plt.show()


```

#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

```python
# MEDIAN STEPS BY MONTH_ENCODED
# your code here

# Set figure size
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 4))

# Task 1: Median steps by month
plt.sca(axes[0])
fitbit2.groupby("Months_encoded")["Steps"].median().plot(kind="bar", color="skyblue")
plt.title("Median Steps by Month")
plt.xlabel("Month")
plt.ylabel("Median Steps")



# MEDIAN STEPS BY WORK_OR_WEEKEND
# your code here

plt.sca(axes[1])
fitbit2.groupby("Work_or_Weekend")["Steps"].median().plot(kind="bar", color="lightcoral")
plt.title("Median Steps on Weekdays vs. Weekends")
plt.xlabel("Day Type")
plt.ylabel("Median Steps")
plt.xticks(ticks=[0, 1], labels=["Weekend", "Weekday"])
```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
# your code here
import os

# Create a folder to save figures
figures_folder = "figures"
os.makedirs(figures_folder, exist_ok=True)

# Define the features for scatter plots
scatter_features = ["Minutes Lightly Active", "Minutes Very Active", "Minutes Sedentary"]


# Set figure size
fig, axes = plt.subplots(nrows=1, ncols=len(scatter_features), figsize=(15, 5))

# Loop to plot scatter plots
for i, feature in enumerate(scatter_features):
    plt.sca(axes[i])
    plt.scatter(fitbit2[feature], fitbit2["Steps"], alpha=0.5)
    plt.title(f"{feature} vs Steps")
    plt.xlabel(feature)
    plt.ylabel("Steps")

# Save figure
plt.tight_layout()
plt.savefig(os.path.join(figures_folder, "scatter_plots.png"))
plt.show()
```

#### Plot the correlation matrix using seaborn.

```python
# Drop rows with missing values in any column
titanic_no_missing = titanic.dropna()

# Exclude non-numeric columns
numeric_columns = titanic_no_missing.select_dtypes(include=['float64', 'int64'])

# Calculate the correlation matrix for numeric features
correlation_matrix = numeric_columns.corr()

# Set up the matplotlib figure
plt.figure(figsize=(12, 10))

# Create a heatmap using Seaborn
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)

# Set title
plt.title('Correlation Matrix of Numerical Features in Titanic Dataset (After Dropping Missing Values)')

# Display the plot
plt.show()
```

#### What are the most correlated features?

```python
# Find the most correlated features (excluding self-correlation and duplicates)
most_correlated_features = correlation_matrix.unstack().sort_values(ascending=False).drop_duplicates().reset_index()
most_correlated_features.columns = ['Feature 1', 'Feature 2', 'Correlation']

# Exclude correlations with themselves
most_correlated_features = most_correlated_features[most_correlated_features['Feature 1'] != most_correlated_features['Feature 2']]

# Display the top correlated features
print(most_correlated_features.head())
```
